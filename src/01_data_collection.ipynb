```python
def split_home_away(series):
    """Convert a Series with Home_ / Away_ stats into a clean DataFrame"""
    data = []
    for ground in ["Home", "Away"]:
        row = {
            "ground": ground,
            "win": series[f"{ground}_wins"],
            "draw": series[f"{ground}_draw"],
            "lose": series[f"{ground}_lose"],
            "scores": series[f"{ground}_goals_for"],
            "conceded": series[f"{ground}_goals_against"]
        }
        data.append(row)
    return pd.DataFrame(data)

 # Home Performance
home_col = ["Home_wins", "Home_draw", "Home_lose", "Home_goals_for", "Home_goals_against"]
home_data = df[home_col]
home_data = home_data.iloc[team_index]

# Away Performance
away_col = ["Away_wins","Away_draw", "Away_lose", "Away_goals_for", "Away_goals_against"]
away_data = df[away_col]
away_data = away_data.iloc[team_index]

df_col = df.columns
df_col = df_col[13:]
data_int = df[df_col]
data_int = data_int.iloc[team_index]

new_df = split_home_away(data_int)
print(new_df)
```

```python
engine = create_engine("sqlite:////Users/thapeloclement/SoccerDashboard/soccer.db", future=True)

def upsert_stats_or_mark_missing(fixture_id, stats_df):
    if stats_df is None or stats_df.empty:
        # mark as no stats
        engine.connect().execute(
            text("INSERT INTO match_stats (fixture_id, has_stats) VALUES (:id, 0) "
                 "ON CONFLICT(fixture_id) DO UPDATE SET has_stats=0"),
            {"id": fixture_id}
        )
        return

for fx_id in missing_fx:
    upsert_stats_or_mark_missing(fx_id, None)
```

```python
from sqlalchemy import create_engine

engine = create_engine(
    "sqlite:////Users/thapeloclement/SoccerDashboard/soccer.db",
    future=True,
    pool_pre_ping=True,
    connect_args={"timeout": 30, "check_same_thread": False},  
)

with engine.begin() as conn:
    conn.exec_driver_sql("PRAGMA journal_mode=WAL;")
    conn.exec_driver_sql("PRAGMA busy_timeout=30000;")


from sqlalchemy import text

SQL_UPSERT = text("""
INSERT INTO match_stats (fixture_id, has_stats)
VALUES (:id, 0)
ON CONFLICT(fixture_id) DO UPDATE SET has_stats=0
""")

def mark_missing_bulk(engine, fx_ids):
    rows = [{"id": i} for i in fx_ids]
    with engine.begin() as conn:         
        conn.execute(SQL_UPSERT, rows)    
```
